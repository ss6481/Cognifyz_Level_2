```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset.csv')
df.head(50)

# 1. Analyze ratings by cuisine
average_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
average_ratings = average_ratings.sort_values(by='Aggregate rating', ascending=False).head(50)

# 2. Identify popular cuisines
vote_counts = df.groupby('Cuisines')['Votes'].sum().reset_index()
vote_counts = vote_counts.sort_values(by='Votes', ascending=False).head(50)

# 3. Merge average ratings with vote counts for comparison
cuisine_stats = pd.merge(average_ratings, vote_counts, on='Cuisines')
top_n = 10  # number of top cuisines to display
top_cuisines = cuisine_stats.nlargest(top_n, 'Votes')

# 4. Visualize the data
plt.figure(figsize=(16, 8))

# Plot average ratings
plt.subplot(1, 2, 1)
plt.barh(cuisine_stats['Cuisines'], cuisine_stats['Aggregate rating'], color='skyblue')
plt.xlabel('Average Rating')
plt.title('Average Rating by Cuisine')

# Plot vote counts
plt.subplot(1, 2, 2)
plt.barh(cuisine_stats['Cuisines'], cuisine_stats['Votes'], color='salmon')
plt.xlabel('Number of Votes')
plt.title('Popularity by Cuisine (Number of Votes)')


plt.show(5)

# Print top cuisines by average rating and votes
print("Top Cuisines by Average Rating:")
print(average_ratings.head(10))

print("\nTop Cuisines by Number of Votes:")
print(vote_counts.head(10))
```
