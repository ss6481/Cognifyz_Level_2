```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Dataset.csv')

print(df.head())

# Distribution of Ratings using histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='red')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

# Average Ratings by Cuisine
average_ratings_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
average_ratings_cuisine = average_ratings_cuisine.sort_values(by='Aggregate rating', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Aggregate rating', y='Cuisines', data=average_ratings_cuisine.head(15), palette='Oranges')
plt.title('Top 15 Cuisines by Average Aggregate Rating')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Cuisine')
plt.show()
```

```python
average_ratings_city = df.groupby('City')['Aggregate rating'].mean().reset_index()
average_ratings_city = average_ratings_city.sort_values(by='Aggregate rating', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Aggregate rating', y='City', data=average_ratings_city.head(15), palette='Oranges')
plt.title('Top 15 Cities by Average Aggregate Rating')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('City')
plt.show()

# Relationship between Features and Ratings
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
sns.pairplot(df[numerical_features])
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

```python
# Box Plot to Compare Distributions by Cuisine
top_cuisines = df['Cuisines'].value_counts().index[:10]
df_top_cuisines = df[df['Cuisines'].isin(top_cuisines)]

plt.figure(figsize=(14, 8))
sns.boxplot(x='Cuisines', y='Aggregate rating', data=df_top_cuisines, palette='Set1')
plt.title('Distribution of Ratings by Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Aggregate Rating')
plt.xticks(rotation=45)
plt.show()

# 6. Box Plot to Compare Distributions by City
top_cities = df['City'].value_counts().index[:10]
df_top_cities = df[df['City'].isin(top_cities)]

plt.figure(figsize=(14, 8))
sns.boxplot(x='City', y='Aggregate rating', data=df_top_cities, palette='Set3')
plt.title('Distribution of Ratings by Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Aggregate Rating')
plt.xticks(rotation=45)
plt.show()
```
